# Ejercicios RDD de pares
1. Inicia las variables SparkSession y SparkContext (0)

```python
from pyspark.sql import SparkSession
sparkSession = SparkSession.builder.getOrCreate()
sc = sparkSession.sparkContext
```

2. Crea un rdd de pares de la forma (palabra, 1) a partir del fichero data/el_quijote.txt (1)

```python
rddq = sc.textFile("data/el_quijote.txt")
rddp = rddq.flatMap(lambda x: x.split())
rdd_pares = rddp.map(lambda palabra: (palabra,1))
#print (rdd_pares.collect())
```

3. A partir del RDD anterior, cuenta el número de ocurrencias de cada palabra (1)

```python
print(rdd_pares.reduceByKey(lambda x,y: x+y).collect())
```

4. Repite los ejercicios 2 y 3 de forma que no se tengan en cuenta los signos de puntación y no se distingan mayúsculas de minúsculas (mapea todo a minúsuculas) (2)

```python
import re

# Cargar el contenido del archivo de texto en un RDD de líneas
lines_rdd = sc.textFile("data/el_quijote.txt")

# Limpiar el texto: convertir a minúsculas y quitar signos de puntuación
cleaned_rdd = lines_rdd.flatMap(lambda line: re.sub(r'[^\w\s]', '', line.lower()).split())

# Crear un RDD de pares (palabra, 1)
word_count_rdd = cleaned_rdd.map(lambda word: (word, 1))

# Contar el número de ocurrencias de cada palabra
word_counts = word_count_rdd.reduceByKey(lambda a, b: a + b)

# Mostrar el resultado
print("Número de ocurrencias de cada palabra:")
print(word_counts.collect())
```

5. A partir de los ficheros data/notas/notas_mates.txt, data/notas/notas_fisica.txt y data/notas/notas_ingles.txt, realiza las siguientes operaciones:
   - a) Crea 3 RDD de pares uno para cada asignatura (notas_mates, notas_ingles, notas_fisica) (0,5)
   - b) Crea un RDD que sea la unión de los 3 anteriores (0,5)
   - c) Muestra la nota más baja de cada alumno (1)
   - d) Muestra la nota media de cada alumno (1)
   - e) Muestra cuantos estudiantes suspenden cada asignatura (1)
   - f) Muestra en qué asignatura suspende más gente (1)
   - g) Obtén un RDD que que asocie a cada alumno con las notas en todas las asignaturas (1)

```python
# a) 
notas_mates_rdd = sc.textFile("data/notas/notas_mates.txt").map(lambda line: (line.split(",")[0], float(line.split(",")[1])))
notas_ingles_rdd = sc.textFile("data/notas/notas_ingles.txt").map(lambda line: (line.split(",")[0], float(line.split(",")[1])))
notas_fisica_rdd = sc.textFile("data/notas/notas_fisica.txt").map(lambda line: (line.split(",")[0], float(line.split(",")[1])))

print("Notas mates:")
print(notas_mates_rdd.collect())
print("\nNotas ingles:")
print(notas_ingles_rdd.collect())
print("\nNotas fisica:")
print(notas_fisica_rdd.collect())
```

```python
# b)
notas_union_rdd = notas_mates_rdd.union(notas_ingles_rdd).union(notas_fisica_rdd)
print("Unión de los tres rdd:\n")
print(notas_union_rdd.collect())

#Otro método para más orden añadiendo la materia a las tubplas
notas_mates_rdd2 = sc.textFile("data/notas/notas_mates.txt").map(lambda line: ("Matematicas", (line.split(",")[0], float(line.split(",")[1]))))
notas_ingles_rdd2 = sc.textFile("data/notas/notas_ingles.txt").map(lambda line: ("Ingles", (line.split(",")[0], float(line.split(",")[1]))))
notas_fisica_rdd2 = sc.textFile("data/notas/notas_fisica.txt").map(lambda line: ("Fisica", (line.split(",")[0], float(line.split(",")[1]))))

combinado = notas_mates_rdd2.union(notas_ingles_rdd2).union(notas_fisica_rdd2)

result_rdd = combinado.flatMap(lambda x: [(x[1][0], (x[0], x[1][1]))])

# Agrupar por clave (nombre del estudiante)
grouped_by_student = result_rdd.groupByKey().map(lambda x: (x[0], tuple(x[1])))
print("\nUnión de los tres rdd ordenado por estudiante:\n")
print(grouped_by_student.collect())
```

```python
# c)
minimo = result_rdd.reduceByKey(lambda a, b: a if a[1] < b[1] else b)
print("Notas mínimas:")
print(minimo.collect())
```

```python
# d)
media = result_rdd.groupByKey().map(lambda x: (x[0], sum(grade[1] for grade in x[1]) / len(x[1])))
print("Medias:")
print(media.collect())
```

```python
#e)
suspenden_mates = notas_mates_rdd.filter(lambda x: x[1] < 5).count()
suspenden_fisica = notas_fisica_rdd.filter(lambda x: x[1] < 5).count()
suspenden_ingles = notas_ingles_rdd.filter(lambda x: x[1] < 5).count()
print(f"Estudiantes que suspenden Matemáticas: {suspenden_mates}")
print(f"Estudiantes que suspenden Física: {suspenden_fisica}")
print(f"Estudiantes que suspenden Inglés: {suspenden_ingles}")
```

```python
# f)
if suspenden_mates > suspenden_fisica and suspenden_mates > suspenden_ingles:
    print(f"La asignatura con más suspendidos es Matemáticas")
elif suspenden_fisica > suspenden_mates and suspenden_fisica > suspenden_ingles:
    print(f"La asignatura con más suspendidos es Física")
else:
    print(f"La asignatura con más suspendidos es Inglés.")
```

```python
# g)
#Esta parte ya la tenía hecha en el apartado b
notas_mates_rdd2 = sc.textFile("data/notas/notas_mates.txt").map(lambda line: ("Matematicas", (line.split(",")[0], float(line.split(",")[1]))))
notas_ingles_rdd2 = sc.textFile("data/notas/notas_ingles.txt").map(lambda line: ("Ingles", (line.split(",")[0], float(line.split(",")[1]))))
notas_fisica_rdd2 = sc.textFile("data/notas/notas_fisica.txt").map(lambda line: ("Fisica", (line.split(",")[0], float(line.split(",")[1]))))

combinado = notas_mates_rdd2.union(notas_ingles_rdd2).union(notas_fisica_rdd2)

result_rdd = combinado.flatMap(lambda x: [(x[1][0], (x[0], x[1][1]))])

# Agrupar por clave (nombre del estudiante)
grouped_by_student = result_rdd.groupByKey().map(lambda x: (x[0], tuple(x[1])))
print("Alumnos y sus notas:")
print(grouped_by_student.collect())
```
